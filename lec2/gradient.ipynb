```python
# data
x_data = [1,2,3]
y_data = [2,4,6]
```

```python
# model
def forward(x,w):
    return x*w
#loss
def loss(x,w,y):
    ypre = forward(x,w)
    return (ypre-y)*(ypre-y)

#gradient
def gradient(w,x,y):
    gra = 2*x**2*w - 2*x*y
    print("gradient: ",gra)
    return gra
# update weight
def update_weight(x,w,rate,y):
    gra = w-rate*(gradient(w,x,y))
    return gra

    
        
    
```

```python
#gradient descent
w = 10
print("before training:",4,forward(4,w))
for j in range(100):
    print("loop: ",j)
    for i in range(len(x_data)):
        w = update_weight(x_data[i],w,0.1,y_data[i])
    print("weight:"+ str(w)+" epoch: "+str(j)+" loss: "+str(loss(w,x_data[i],y_data[i])))
print("after training:",4,forward(4,w))
        
```

```python
# model
def forward2(x,w2,w1,b):
    return x**2*w2+x*w1+b
#loss
def loss2(x,w2,w1,b,y):
    ypre = forward2(x,w2,w1,b)
    return (ypre-y)*(ypre-y)

#gradient
def gradient_w2(x,w2,w1,b,y):
    gra = (2*x**2)*(x**2*w2 + x*w1 + b - y)
    return gra

def gradient_w1(x,w2,w1,b,y):
    gra = (2*x)*(x**2*w2 + x*w1 + b - y)
    return gra

def gradient_b(x,w2,w1,b,y):
    gra = 2*(x**2*w2 + x*w1 + b - y)
    return gra
# update weight
def update_weight2(x,w2,w1,b,rate,y):
    w2_new = w2-rate*(gradient_w2(x,w2,w1,b,y))
    w1_new = w1-rate*(gradient_w1(x,w2,w1,b,y))
    b_new = b-rate*(gradient_b(x,w2,w1,b,y))
    return w2_new,w1_new,b_new
#     w2 = w2-rate*(gradient_w2(x,w2,w1,b,y))
#     w1 = w1-rate*(gradient_w1(x,w2,w1,b,y))
#     b = b-rate*(gradient_b(x,w2,w1,b,y))
#     return w2,w1,b
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
x_data2 = np.arange(-14,14,0.1)
y_data2 = [x**2+5*x+1+0.1 for x in x_data2]
plt.scatter(x_data2, y_data2)
plt.show()
```

gradient descent to find w1,w2,b

```python
w2 = 5
w1 = 5
b = 5
print("before training:",15,forward2(15,w2,w1,b))
for j in range(5000):
    list_loss = []
    print(" w2:",w2)
    print(" w1:"+ str(w1))
    print(" b:"+ str(b))
    
    for i in range(len(y_data2)):
        list_loss.append(loss2(x_data2[i],w2,w1,b,y_data2[i]))
        w2,w1,b = update_weight2(x_data2[i],w2,w1,b,0.00005,y_data2[i])
    print("epoch: "+ str(j)+", mean loss: "+str(np.mean(list_loss)))
    
print("final\n w1: "+str(w1) +" w2: "+str(w2)+" b: "+str(b))
print("predict (after training)","15", forward2(15,w2,w1,b))    

    
```
