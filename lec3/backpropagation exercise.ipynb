```python
def forward2(x,w,b):
    return x*w + b

def loss2(x,w,b,y):
    return (x*w + b - y)**2

def gradient_loss_w2(x,w,b,y):
    return 2*x*(x*w + b - y)

def gradient_loss_b(x,w,b,y):
    return 2*(x*w + b - y)

def update_weight2(x,w,b,y,rate):
    w = w - rate*gradient_loss_w2(x,w,b,y)
    b = b - rate*gradient_loss_b(x,w,b,y)
    return w,b
```

```python
x_data = [1,2,3]
y_data = [3,5,7]
w = 0
b = 0
for j in range(1000):
    for i in range(len(x_data)):
        w,b = update_weight2(x_data[i],w,b,y_data[i],0.1)
        print("\tgrad: ", x_data[i], y_data[i], gradient_loss_w2(x_data[i],w,b,y_data[i]))
        l = loss2(x_data[i],w,b,y_data[i])
    print("progress:", j, "w=", w, "b=", b, "loss=", l)
print("predict (after training)", " x = ",4," y = ", forward2(4,w,b))
```
