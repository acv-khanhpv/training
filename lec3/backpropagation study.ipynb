```python
def forward(x,w):
    return x*w

def loss(x,w,y):
    return (x*w-y)**2

def gradient_loss_w(x,w,y):
    return 2*x*(x*w-y)

def update_weight(x,w,y,rate):
    return w - rate*gradient_loss_w(x,w,y)
```

```python
gradient_loss_w(2,1,4)
```

```python
x_data = [1,2,3]
y_data = [2,4,6]
w = 0
for j in range(100):
    for i in range(len(x_data)):
        w = update_weight(x_data[i],w,y_data[i],0.1)
        print("\tgrad: ", x_data[i], y_data[i], gradient_loss_w(x_data[i],w,y_data[i]))
        l = loss(x_data[i],w,y_data[i])
    print("progress:", j, "w=", w, "loss=", l)
```
