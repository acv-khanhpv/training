```python
x = [1,2,3]
y = [2,4,6]
```

```python
sumloss = 0
```

```python
def loss(x,w,y):
    return (x*w-y)*(x*w-y)
loss(2,4,2)
```

```python
def meanloss(data_x,data_y,w):
    sumloss = 0
    for i in range(len(data_x)):
        sumloss+=loss(x[i],w,y[i])
        print(sumloss)
    mean = sumloss/len(data_x)
    print (mean)
    return mean
meanloss(x,y,1)
```

```python
import pandas as pd
data = pd.read_csv('data.csv')
```

```python
row = [""]
for i in range(5):
    row.append(meanloss(x,y,i))
row
```
