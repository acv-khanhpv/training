```python
label = data_train['Survived'].astype(np.float32)
```

```python
data_train = data_train.drop(['Ticket','Name', 'Ticket', 'PassengerId','Survived', 'Cabin'], axis = 1)
data_train.head()
```

```python
def myfillna(series):
    if series.dtype is pd.np.dtype(float):
        return series.fillna(0)
    elif series.dtype is pd.np.dtype(object):
        return series.fillna('UNKNOW')
    else:
        return series
```

```python
data_train = data_train.apply(myfillna)
```

```python
data_train = pd.concat([data_train.drop('Sex', axis=1), pd.get_dummies(data_train['Sex'])], axis=1)
data_train = pd.concat([data_train.drop('Embarked', axis=1), pd.get_dummies(data_train['Embarked'])], axis=1)
data_train = pd.concat([data_train.drop('Pclass', axis=1), pd.get_dummies(data_train['Pclass'])], axis=1)
data_train = pd.concat([data_train.drop('SibSp', axis=1), pd.get_dummies(data_train['SibSp'])], axis=1)
data_train = pd.concat([data_train.drop('Parch', axis=1), pd.get_dummies(data_train['Parch'])], axis=1)
```
