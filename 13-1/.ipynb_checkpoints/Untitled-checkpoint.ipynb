```python
x_data = [2.1, 4.2, 3.1, 3.3]
y_data = [0.,1.,0.,1.]
```

```python
import numpy as np
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def forward(x,w,b):
    return x*w + b
```

```python
def gradient_sigmoid_w(x,w,b,y):
    return -(y - sigmoid(forward(x,w,b)))*x
```

```python
def gradient_sigmoid_b(x,w,b,y):
    return -(y - sigmoid(forward(x,w,b)))
```

```python

def train(rate, num_epochs):
    w = 100
    b = -100
    for _ in range(num_epochs):
        for i in range(len(x_data)):
            w = w - rate*gradient_sigmoid_w(x_data[i],w,b,y_data[i])
            b = b - rate*gradient_sigmoid_b(x_data[i],w,b,y_data[i])
            
        print("w: ",w, "b: ", b)
    return w,b
```

```python
w,b = train(0.1, 10000)
```

```python
sigmoid(forward(3.3,w,b))
```

```python
def forward_2(x1,w1,x2,w2,b):
    return x1*w1+x2*w2+b
```

```python
def gradient_w1(x1,w1,x2,w2,b,y):
    return -(y-(sigmoid(forward_2(x1,w1,x2,w2,b)))*x1)

def gradient_w2(x1,w1,x2,w2,b,y):
    return -(y-(sigmoid(forward_2(x1,w1,x2,w2,b)))*x2)

def gradient_b(x1,w1,x2,w2,b,y):
    return -(y-(sigmoid(forward_2(x1,w1,x2,w2,b))))
```
